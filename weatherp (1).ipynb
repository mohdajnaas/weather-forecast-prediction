```python
#modules/libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, classification_report
```

```python
#upload dataset
from google.colab import files

uploaded = files.upload()
```

```python
#read and display dataset
import io

df = pd.read_csv(io.BytesIO(uploaded['seattle-weathernodate.csv']))
print(df)
```

```python
#check data type of contents
df.dtypes
```

```python
#encord strings to numericals
Numerics = LabelEncoder()
```

```python
#drop the target variable and make it as nerw frame
inputs = df.drop('weather',axis = 'columns')
target = df['weather']
target
```

```python
#applying Gaussian Naive bayes
Classifier = GaussianNB()
Classifier.fit(inputs,target)
```

```python
Classifier.score(inputs,target)
```
